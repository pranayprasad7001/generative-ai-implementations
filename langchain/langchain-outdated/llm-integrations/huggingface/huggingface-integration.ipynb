### Huggingface Integrations

```python
import os
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace

hf_token = os.getenv("HF_TOKEN")
```

```python
repo_id = "deepseek-ai/DeepSeek-V4-Pro"

model = HuggingFaceEndpoint(
    repo_id=repo_id,
    max_new_tokens=150,
    temperature=0.9,
    task="text-generation",
    huggingfacehub_api_token=hf_token, 
)

llm = ChatHuggingFace(llm=model, verbose=True)
```

```python
# Pass a string instead of a dictionary
response = llm.invoke("What is machine learning")
print(response.content)
```

```python
from langchain_classic import PromptTemplate, LLMChain

template = """
Question: {question}
Answer: Let think step by step
"""

prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(llm=llm, prompt=prompt)
llm_chain.invoke("who won the cricket world cup 2011")
```
