### Stream and Batch

```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# Load Environment Variables 
load_dotenv()
```

#### Streamming 

```python
model = init_chat_model(model="google_genai:gemini-2.5-flash-lite")
```

```python
response = model.invoke("Write me a 200 words paragraph on AI")
response.content
```

```python
for chunk in model.stream("Write me a 200 words paragraph on AI"):
    print(chunk.text, end="|", flush=True)
```

```python
for chunk in model.stream("Why parrots have feathers?"):
    print(chunk.text, end="|", flush=True)
```

#### Batch

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is quantam computing?"
    ]
)

for response in responses:
    print(response)
```

```python
# Using config parameter to limit parallel calls at once
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is quantam computing?",
        "How do cars run on roads?",
        "How do GPUs work?",
        "How to navigate using GPS?"
    ],
    config={'max_concurrency': 5,}
)

for response in responses:
    print(response)
```
