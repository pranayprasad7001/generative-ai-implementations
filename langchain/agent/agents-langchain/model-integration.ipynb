### Model Integration with OpenAI, Google Gemini and GROQ

```python
from dotenv import load_dotenv

# Load Environment Variables 
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="groq:llama-3.1-8b-instant") 
model
```

```python
# Inoke a Model
response = model.invoke("Hi, How are you?")
response.content
```

### Model Integrations - Specific to Model Providers

```python
# Lanchain Based Model Integration from Different Providers
model = init_chat_model(model="google_genai:gemini-2.5-flash-lite")
response = model.invoke("Why do parrots talk?")
response.content
```

```python
# OpenAI Model Integration 
from langchain_openai import ChatOpenAI

openai_model = ChatOpenAI(model="gpt-5.5")
response_openai = openai_model.invoke("What is Langsmith?")
response_openai.content
```

```python
# Groq Model Integration
from langchain_groq import ChatGroq

model_groq = ChatGroq(model="openai/gpt-oss-120b")
response_groq = model_groq.invoke("What is LangServe?")
response_groq.content
```

```python
# Google Model Integration
from langchain_google_genai import ChatGoogleGenerativeAI

model_goo_genai = ChatGoogleGenerativeAI(model="gemini-3.5-flash")
response_goo_genai = model_goo_genai.invoke("What is UV Manager for venvs?")
response_goo_genai.text
```
